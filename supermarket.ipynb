# Visualizaciones

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Product line')
plt.title('Frecuencia de cada lnea de producto')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Product line', hue='Gender')
plt.title('Distribucin de Lnea de Producto por Gnero')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Ventas por producto
sales_per_product = data.groupby('Product line')['Total'].sum()

sales_per_product.plot(kind='bar')
plt.title('Ventas por Producto')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

# Analisis general

```python
sales_per_product = data.groupby('Product line')['Total'].sum().sort_values(ascending=False)
sales_per_product
```

```python
sales_per_city = data.groupby('City')['Total'].sum().sort_values(ascending=False)
sales_per_city
```

```python
data['Date'] = pd.to_datetime(data['Date'])
sales_over_time = data.groupby('Date')['Total'].sum()

sales_over_time.plot(kind='line', figsize=(10,6))
plt.title('Ventas a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas Totales')
plt.grid(True)
plt.show()
```

```python
df = data.drop(columns=['Invoice ID', 'cogs', 'gross margin percentage',"gross income","Tax 5%","Time"])
df.head()
```

```python
customer_type_segmentation = df.groupby('Customer type')['Total'].sum()
customer_type_segmentation
```

```python
gender_segmentation=df.groupby('Gender')['Total'].sum()
gender_segmentation
```

```python
payment_method_segmentation=df.groupby('Payment')['Total'].sum()
payment_method_segmentation
```
