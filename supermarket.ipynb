# Visualizaciones

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Product line')
plt.title('Frecuencia de cada línea de producto')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Product line', hue='Gender')
plt.title('Distribución de Línea de Producto por Género')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Ventas por producto
sales_per_product = data.groupby('Product line')['Total'].sum()

sales_per_product.plot(kind='bar')
plt.title('Ventas por Producto')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

```python
sales_per_product = data.groupby('Product line')['Total'].sum().sort_values(ascending=False)
sales_per_product
```

```python
sales_per_city = data.groupby('City')['Total'].sum().sort_values(ascending=False)
sales_per_city

sales_per_city.plot(kind='bar')
plt.title('Ventas por Ciudad')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
sales_over_time = data.groupby('Date')['Total'].sum()

sales_over_time.plot(kind='line', figsize=(10,6))
plt.title('Ventas a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas Totales')
plt.grid(True)
plt.show()
```

```python
df = data.drop(columns=['Invoice ID', 'cogs', 'gross margin percentage',"gross income","Tax 5%","Time"])
df.head()
```

```python
customer_type_segmentation = df.groupby('Customer type')['Total'].sum()
customer_type_segmentation

customer_type_segmentation.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Distribución de Ventas por Tipo de Cliente')
plt.ylabel('')
plt.show()
```

```python
gender_segmentation=df.groupby('Gender')['Total'].sum()
gender_segmentation

gender_segmentation.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Distribución de Ventas por Género')
plt.ylabel('')
plt.show()
```

```python
payment_method_segmentation=df.groupby('Payment')['Total'].sum()
payment_method_segmentation

payment_method_segmentation.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Distribución de Ventas por Método de Pago')
plt.ylabel('')
plt.show()
```

```python
Sales_count=df.groupby(["City","Product line",]).size().reset_index(name="Sales_count")

most_sold_product=Sales_count.sort_values(by="Sales_count",ascending=False).groupby("City").first().reset_index()
most_sold_product



```

```python
most_sold_product.plot(kind='bar', x='City', y='Sales_count', figsize=(10,6))
plt.title('Productos más vendidos por ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad de ventas')
plt.grid(True)

for index, row in most_sold_product.iterrows():
    plt.text(index, row['Sales_count'], str(row['Product line']), ha='center', va='bottom')

plt.show()
```

```python
payment_per_city=df.groupby(["City","Payment"]).size().reset_index(name="Payment_count")

most_used_payment=payment_per_city.sort_values(by="Payment_count",ascending=False).groupby("City").first().reset_index()
most_used_payment
```

```python
most_used_payment.plot(kind='bar', x='City', y='Payment_count', figsize=(10,6))
plt.title('Métodos de pago más utilizados por ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad de ventas')
plt.grid(True)

for index, row in most_used_payment.iterrows():
    plt.text(index, row['Payment_count'], str(row['Payment']), ha='center', va='bottom')

plt.show()
```

```python
customer_per_city=df.groupby(["City","Customer type"]).size().reset_index(name="Customer_count")

most_used_customer=customer_per_city.sort_values(by="Customer_count",ascending=False).groupby("City").first().reset_index()
most_used_customer
```

```python
most_used_customer.plot(kind='bar', x='City', y='Customer_count', figsize=(10,6))
plt.title('Tipo de cliente más utilizado por ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad de ventas')
plt.grid(True)

for index, row in most_used_customer.iterrows():
    plt.text(index, row['Customer_count'], str(row['Customer type']), ha='center', va='bottom')

plt.show()
```

```python
sex_por_city=df.groupby(["City","Gender"]).size().reset_index(name="Gender_count")

most_used_sex=sex_por_city.sort_values(by="Gender_count",ascending=False).groupby("City").first().reset_index()
most_used_sex
```

```python
most_used_sex.plot(kind='bar', x='City', y='Gender_count', figsize=(10,6))
plt.title('Género más utilizado por ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad de ventas')
plt.grid(True)

for index, row in most_used_sex.iterrows():
    plt.text(index, row['Gender_count'], str(row['Gender']), ha='center', va='bottom')

plt.show()
```

```python
clientes_per_sex=df.groupby(["Gender","Customer type"]).size().reset_index(name="Customer_count")

most_used_customer_sex=clientes_per_sex.sort_values(by="Customer_count",ascending=False).groupby("Gender").first().reset_index()
most_used_customer_sex

clientes_per_sex.plot(kind='bar', x='Gender', y='Customer_count', figsize=(10,6))
plt.title('Tipo de cliente más utilizado por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de ventas')
plt.grid(True)

for index, row in most_used_customer_sex.iterrows():
    plt.text(index, row['Customer_count'], str(row['Customer type']), ha='center', va='bottom')

plt.show()

clientes_per_sex
```

# Hallazgos General y separados por ciudad

**General**
- **Comportamiento de compra por género**: Las mujeres parecen ser las principales consumidoras en este contexto, lo que podría indicar que los productos o la experiencia de compra atraen más a las mujeres.
- **Método de pago preferido**: El efectivo es el método de pago más utilizado. Esto podría sugerir que los clientes prefieren pagar en efectivo o que hay una oportunidad para aumentar las ventas promoviendo más los pagos digitales.
- **Tipo de cliente**: Los miembros son más frecuentes que los clientes normales, lo que podría indicar que el programa de membresía es atractivo para los clientes y podría ser una buena área para seguir invirtiendo.
- **Ventas por ciudad**: Naypyitaw es la ciudad con más ventas, lo que podría sugerir que esta ciudad tiene una gran demanda de los productos o que las estrategias de marketing están funcionando bien allí.
- **Línea de producto más vendida**: Los alimentos y bebidas son los productos más vendidos. Esto podría indicar que estos productos son muy populares entre los clientes y podrían ser el foco principal para futuras estrategias de marketing y ventas.
- Las mujeres se unen mas a la membresia que los hombres

**Por ciudad**
- En Mandalay los hombres compran mas que las mujeres,el tipo de cliente es mas frecuente el cliente noirmal que miembros.El pago mas usado es Ewallet y su producto mas vendido es sport and travel.
- En Naypytaw las mujeres compran mas, el cliente mas frecuente son los miembros, el metodo de pago mas usado es cash y el producto mas vendido es "Food and beverages"
- Yangon los hombres compran mas, los clientes mas frecuentes son los normales, el metodo de pago mas usado es Ewallet y el producto mas vendido es "Home and lifestyle"
