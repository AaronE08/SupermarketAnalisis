# Visualizaciones

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Product line')
plt.title('Frecuencia de cada línea de producto')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Product line', hue='Gender')
plt.title('Distribución de Línea de Producto por Género')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Ventas por producto
sales_per_product = data.groupby('Product line')['Total'].sum()

sales_per_product.plot(kind='bar')
plt.title('Ventas por Producto')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

# Analisis general

```python
sales_per_product = data.groupby('Product line')['Total'].sum().sort_values(ascending=False)
sales_per_product
```

```python
sales_per_city = data.groupby('City')['Total'].sum().sort_values(ascending=False)
sales_per_city
```

```python
data['Date'] = pd.to_datetime(data['Date'])
sales_over_time = data.groupby('Date')['Total'].sum()

sales_over_time.plot(kind='line', figsize=(10,6))
plt.title('Ventas a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas Totales')
plt.grid(True)
plt.show()
```

```python
df = data.drop(columns=['Invoice ID', 'cogs', 'gross margin percentage',"gross income","Tax 5%","Time"])
df.head()
```

```python
customer_type_segmentation = df.groupby('Customer type')['Total'].sum()
customer_type_segmentation
```

```python
gender_segmentation=df.groupby('Gender')['Total'].sum()
gender_segmentation
```

```python
payment_method_segmentation=df.groupby('Payment')['Total'].sum()
payment_method_segmentation
```

# Hallazgos

- **Comportamiento de compra por género**: Las mujeres parecen ser las principales consumidoras en este contexto, lo que podría indicar que los productos o la experiencia de compra atraen más a las mujeres.
- **Método de pago preferido**: El efectivo es el método de pago más utilizado. Esto podría sugerir que los clientes prefieren pagar en efectivo o que hay una oportunidad para aumentar las ventas promoviendo más los pagos digitales.
- **Tipo de cliente**: Los miembros son más frecuentes que los clientes normales, lo que podría indicar que el programa de membresía es atractivo para los clientes y podría ser una buena área para seguir invirtiendo.
- **Ventas por ciudad**: Naypyitaw es la ciudad con más ventas, lo que podría sugerir que esta ciudad tiene una gran demanda de los productos o que las estrategias de marketing están funcionando bien allí.
- **Línea de producto más vendida**: Los alimentos y bebidas son los productos más vendidos. Esto podría indicar que estos productos son muy populares entre los clientes y podrían ser el foco principal para futuras estrategias de marketing y ventas.
